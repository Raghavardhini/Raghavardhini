```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/BKB_WaterQualityData_2020084.csv")
data
```

```python
plt.scatter(data['WT'],data['Year'],color='red')
plt.xlim(1,35)
plt.ylim(1940,2000)
plt.xlabel("Water Temperature")
plt.ylabel("Year")
plt.show()
```

```python
x = data.iloc[:,5:7]
x
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python

kmeans = KMeans(3)
```

```python
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = data.copy()
```

```python
data_with_clusters
```

```python
data_with_clusters['C'] = identified_clusters
plt.scatter(data_with_clusters['WD'],data_with_clusters['WD'],c=data_with_clusters['C'],cmap='rainbow')
```

```python
data_with_clusters
```

```python
individual_clustering_score=[]
for i in range(1,4):
    kmeans=KMeans(n_clusters=i,init='random',random_state=42)
    kmeans.fit(x)
    individual_clustering_score.append(kmeans.inertia_)
plt.plot(range(1,4),individual_clustering_score)
plt.title("WaterQuality")
plt.show()
```
