```python
import pandas as pd
```

```python
df = pd.read_csv("/content/STUDENT.csv")
df
```

```python
df = df.fillna(2)
df
```

```python
df = df.dropna()
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = pd.get_dummies(df, columns=['DEPARTMENT'])
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df["SALARY"] = scaler.fit_transform(df["SALARY"].values.reshape(-1, 1))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df['ID'], df['SALARY'], color='red', width=0.4)
plt.xlabel("ID")
plt.ylabel("SALARY")
plt.show()
```
